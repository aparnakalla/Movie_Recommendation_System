```python
avg_rating=df2['vote_average'].mean()
```

```python
min_votes=df2['vote_count'].quantile(0.9)
min_votes
```

```python
list_movies=df2.copy().loc[df2['vote_count']>=min_votes]
list_movies
```

```python
def weighted_rating(x,c=avg_rating,m=min_votes):
    v=x['vote_count']
    R=x['vote_average']
    return(v/(v+m)*R)+(v/(v+m)*c)
```

```python
list_movies['score']=list_movies.apply(weighted_rating,axis=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(stop_words='english')

df2['overview']=df2['overview'].fillna('')

tfidf_matrix=tfidf.fit_transform(df2['overview'])

tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel 

cosine_sim=linear_kernel(tfidf_matrix,tfidf_matrix)
```

```python
cosine_sim
```

```python
cosine_sim.shape
```

```python
indices=pd.Series(df2.index,index=df2['title_x']).drop_duplicates()
```

```python
def get_recommendations(title,cosine_sim=cosine_sim):
    
    idx=indices[title]
    
    
    sim_scores=list(enumerate(cosine_sim[idx]))
    
    
    sim_scores=sorted(sim_scores,key=lambda x:x[1], reverse=True)
    
    
    sim_scores=sim_scores[1:11]
    
    
    movie_indices=[i[0] for i in sim_scores]
    
    
    return df2['title_x'].iloc[movie_indices]
```

```python
get_recommendations('Avatar')
```
